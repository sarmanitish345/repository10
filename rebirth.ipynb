```python
def search(toFind):
    if toFind == "^^":
        return None
    elif toFind=="$$":
        return None
    elif ord(toFind[0])<200:
        return None
    else:
        return df2.loc[toFind]
search(list[2])
```

```python
def loopVariable(str1,str2,str3):
    if ord(str1[0])<200:
        ith=2
    else:
        ith=11
    if ord(str2[0])<200:
        jth=2
    else:
        jth=11
    if ord(str3[0])<200:
        kth=2
    else:
        kth=11
    return [ith,jth,kth]
```

```python
def printPermutation(row):
    cleanString=removeSpace(row)
    list=cleanString.split("_")
    loopVar=loopVariable(list[1],list[2],list[3])
    for i in range(1,loopVar[0]):
        for j in range(1,loopVar[1]):
            for k in range(1,loopVar[2]):
                obj1=search(list[1])
                obj2=search(list[2])
                obj3=search(list[3])
                
                if(obj1 is None):
                    first=list[1]
                else:
                    first=obj1[i]
                
                if(obj2 is None):
                    second=list[2]
                else:
                    second=obj2[j]
                    
                if(obj3 is None):
                    third=list[3]
                else:
                    third=obj3[k]
                    
                print(first+" "+second+" "+third)
```
